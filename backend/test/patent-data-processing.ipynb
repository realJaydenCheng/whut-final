```python
import os

import pandas as pd

DIR = "C:/Users/realj/Desktop/Projects/whut-final/backend/var/patent"
```

```python
file_names = os.listdir(DIR)
file_names
```

```python
data = pd.concat([
    pd.read_csv(DIR+"/"+file_name, encoding="gb18030")
    for file_name in file_names
    if file_name.endswith(".csv")
])
```

```python
tgt_data = data[[
    "placard_num", "apply_date", "pat_name",
    "pat_agent", "pat_applicant", "legal_status",
    "pat_validity", "pat_type", "pat_summary"
]]
```

```python
tgt_data.head(3)
```

```python
import numpy as np
dates = [f"20{x}-10-20" for x in range(10, 24) ]
tgt_data = tgt_data.rename(columns={
    "placard_num":"编号",
    "apply_date":"申请日期",
    "pat_name":"名称",
    "pat_agent":"代理",
    "pat_applicant":"应用",
    "legal_status":"状态",
    "pat_validity":"验证",
    "pat_type":"类型",
    "pat_summary":"概述",
})
tgt_data["分析"] = np.random.choice(dates, size=len(tgt_data), replace=True)
tgt_data.head()
```

```python
len(tgt_data)
```

```python

files_out_cnt = 6

one_file_size = (len(tgt_data) // files_out_cnt)
for i in range(files_out_cnt + 1):
    tgt_data.iloc[i*one_file_size: (i+1)*one_file_size].to_excel(DIR+"/"+f"{i}patent.xlsx", index=False)
```
