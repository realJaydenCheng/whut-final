```python
import matplotlib.pyplot as plt
import matplotlib_inline

matplotlib_inline.config.InlineBackend.figure_format = "svg"
plt.rcParams['font.sans-serif'] = 'Microsoft YaHei'
plt.rcParams['axes.unicode_minus'] = False
```

```python
value_counts = value_counts.drop("")
```

```python

value_counts.plot()  # 你也可以选择其他类型的图
# 设置 x 轴标签旋转 45 度
plt.xticks(rotation=45)
```

```python
threshold = 100

def replace_cate_label(x: str):
    value = 0
    if x in value_counts.index:
        value = value_counts[x]
    if value < threshold:
        return ""
    return x

data_frame["所属专业大类"] = data_frame["所属专业大类"].apply(replace_cate_label)
data_frame
```

```python
value_counts = data_frame["所属专业大类"].value_counts()
print(
    list(zip(value_counts.index, list(value_counts)))
)
value_counts = value_counts.drop("")

value_counts.plot()
plt.xticks(rotation=30)

```

```python
"""
data_frame: pd.DataFrame
我想将data_frame按照年份保存为xlsx格式。
具体来说，根据“立项时间”字段的前四个字符来作为年份，
保存的时候文件路径为TGT+“{年份字符}.xlsx”，index=false，dtype=str
"""

# 假设 TGT 是目标文件夹的路径
# 检查 '立项时间' 是否已经是 datetime 类型，如果不是，则先转换
# if not pd.api.types.is_datetime64_any_dtype(data_frame['立项时间']):
#     data_frame['立项时间'] = pd.to_datetime(data_frame['立项时间'], format='%Y-%m-%d', errors='coerce')

# 从 '立项时间' 提取年份
data_frame['temp'] = data_frame['立项时间'].apply(lambda x: x[:4])

# 按年份分组并保存每个组为一个 xlsx 文件
for year, group in data_frame.groupby('temp'):
    file_path = TGT + f"{year}.xlsx"
    group = group.drop(columns=["temp"])
    group.to_excel(file_path, index=False)

print("文件已按年份保存在指定目录。")
```
