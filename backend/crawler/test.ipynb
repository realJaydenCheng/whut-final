```python

trs = items_list_2023_soup.select('tr:not(thead tr)')
data_list = []

for tr in trs:
    # 提取编号
    number = tr['id'].split('_')[-1]

    # 提取代码
    code = tr.find_all('td')[0].get_text(strip=True)

    # 提取成员
    members = tr.find_all('td')[2].get_text(strip=True).split('、')

    # 提取级别
    level = tr.find_all('td')[3].get_text(strip=True)

    # 提取指导老师
    teachers = tr.find_all('td')[4].get_text(strip=True).split('、')

    # 提取学校
    school = tr.find_all('td')[5].get_text(strip=True)

    # 构建字典并添加到列表中
    data_list.append({
        "number": number,
        "code": code,
        "members": members,
        "level": level,
        "teachers": teachers,
        "school": school
    })

# 输出结果
data_list
```

```python
# labels tag: //div/label/label
# about following-sibling: https://blog.csdn.net/weixin_44090305/article/details/92185565
# values tag: //div/label/label/../following-sibling::div

labels = [
    item for item in
    items_detail_soup.select('div > label > label')
    if (
        "指导教师" not in item.text.strip() and
        "项目成员" not in item.text.strip()
    )
]
values = [
    label.parent.find_next_sibling('div')
    for label in labels
]
print([tag.text.strip() for tag in labels])
print([tag.text.strip() for tag in values])
```
